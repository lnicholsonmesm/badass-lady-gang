```python
# SQLAlchemy
from sqlalchemy import create_engine
import pandas as pd

# Imports the methods needed to abstract classes into tables
from sqlalchemy.ext.declarative import declarative_base

# Allow us to declare column types
from sqlalchemy import Column, Integer, String, Float, insert
from sqlalchemy.orm import sessionmaker
```

```python
engine = create_engine("sqlite:///Internet_Data_Final.db", echo=False)
conn = engine.connect()
```

```python
df = pd.read_csv('CompInt_Race_TblS2802.csv', encoding="utf-8", error_bad_lines = False)
df.to_sql('Race_Data_final', conn, if_exists='replace', index=False)
```
